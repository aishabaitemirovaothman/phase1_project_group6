```python
import pandas as pd
import numpy as np
```

```python
#display dataframe
movie_type= pd.read_csv('data/tmdb.movies.csv.gz')
display(movie_type)
display(movie_type.info())
```

```python
movie_type.sort_values(by='popularity', ascending=False  )
```

```python
movie_type= movie_type.loc[movie_type['original_language'] == "en"]
movie_type
```

```python
movie_type= movie_type.loc[movie_type['vote_count'] >= 5000]
movie_type
```

```python
movie_type["genre_ids"].iloc[0]
```

```python
movie_type.genre_ids = movie_type.genre_ids.map(lambda x:eval(x))
```

```python
movie_type.explode("genre_ids")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks", color_codes=True)
```

```python
#Create a violin plot to display the relationship between vote averages and genres
g = sns.catplot(x=movie_type["genre_ids"].astype("category") , y="vote_average", kind="violin", inner=None, data=movie_type)
#sns.swarmplot(x="genre_ids", y="vote_average", color="k", size=3, data=movie_type, ax=g.ax)
```
